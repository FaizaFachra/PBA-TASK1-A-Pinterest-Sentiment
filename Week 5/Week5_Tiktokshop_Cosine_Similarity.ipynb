```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
# 1. Dataset dari slide Stanford
corpus = [
    "Tokopedia dan TikTok Shop bersama KemenUMKM melatih juragan UMKM agar bisa jualan nyaman lewat program Creators Lab.", # Kalimat 1
    "Program Creators Lab ini dihadirkan untuk meningkatkan kompetensi UMKM dalam memanfaatkan platform digital seperti Tokopedia dan TikTok Shop.", # Kalimat 2
    "Melalui kolaborasi dengan KemenUMKM, Tokopedia dan TikTok Shop berkomitmen mendukung pertumbuhan UMKM melalui inovasi Creators Lab." # Kalimat 3
]

# 2. Representasi Vektor (Hanya fokus pada kata 'umkm' dan 'creators')
vectorizer = CountVectorizer(vocabulary=['umkm', 'creators'])
X = vectorizer.fit_transform(corpus)
```

```python
# 3. Hitung Cosine Similarity
# Membandingkan semua dokumen dengan Dokumen 0
cos_sim = cosine_similarity(X[0:1], X)

# Tampilkan Hasil
df = pd.DataFrame(X.toarray(), columns=['umkm', 'creators'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()
print(df)
```

```python
documents = [
    "Tokopedia dan TikTok Shop bersama KemenUMKM melatih juragan UMKM agar bisa jualan nyaman lewat program Creators Lab.",
    "Program Creators Lab ini dihadirkan untuk meningkatkan kompetensi UMKM dalam memanfaatkan platform digital seperti Tokopedia dan TikTok Shop.",
    "Melalui kolaborasi dengan KemenUMKM, Tokopedia dan TikTok Shop berkomitmen mendukung pertumbuhan UMKM melalui inovasi Creators Lab."
]
```

```python
import pandas as pd
from collections import Counter
tf = pd.DataFrame(
    [Counter(doc.lower().replace('.', '').split()) for doc in documents],
).fillna(0).astype(int)

# Menampilkan matriks
print("Matriks Term Frequency (TF):")
print(tf)
```

```python
full_cos_sim_matrix = cosine_similarity(X)

# Display the full matrix as a DataFrame for better readability
cos_sim_df = pd.DataFrame(full_cos_sim_matrix, index=[f'Doc {i}' for i in range(len(corpus))], columns=[f'Doc {i}' for i in range(len(corpus))])
print("Full Cosine Similarity Matrix:")
print(cos_sim_df)
```
